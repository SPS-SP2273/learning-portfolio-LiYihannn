## Exercise 1 :  Fill in the blanks

```python
# Some data for plotting
# I have tried my best to figure out the color of the line alr, they look the same for me...
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, xerr=x_err, yerr=y_err,
             color="#e34a33", linestyle="dashed",
             marker="o", markerfacecolor="yellow",
             ecolor="black",capsize=5,
             label = "Y values with errors")
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='lower right')             
plt.show(block=False)
```

## Exercise 2 :  A simple mathematical plot

```python
# Q1
import numpy as np
x=np.arange(0,50.1,0.1)
y=np.exp(-x/10) * np.sin(x)
```

```python
# Q2
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,50.1,0.1)
y=np.exp(-x/10) * np.sin(x)

plt.plot(x, y)
plt.xlabel('x',color="darkblue")
plt.ylabel('y',color="darkblue")
plt.ylabel("y", rotation=0)
plt.rc('axes', labelsize=15) 
plt.title(r'$y=e^{-x/10}sin(x)$',fontsize=20)
plt.grid(alpha=.25) 
plt.ylim([-1,1])
plt.show()
```
