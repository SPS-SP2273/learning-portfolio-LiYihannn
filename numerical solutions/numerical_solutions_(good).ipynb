## Exercise 1 :  The Lorentz Attaractor

```python
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import matplotlib


x_range=[-20,21]
y_range=[-25,31]
z_range=[0,41]
dt = 0.001
time = 0
time_max = 50
intial_x = 0
intial_y = 1
intial_z = 0
x = intial_x
y = intial_y
z = intial_z
data = []


# Lotka-Volterra equations
def dx_dt(x, y):
    return 10*(y-x)

def dy_dt(x, y, z):
    return x*(28-z)-y

def dz_dt(x, z):
    return x*y-8/3*z

# generating data set 
while time < time_max:
    data.append([time, x, y, z])

    x_new = x + dx_dt(x, y)*dt
    y_new = y + dy_dt(x, y, z)*dt
    z_new = z + dz_dt(x, z)*dt    

    x = x_new
    y = y_new
    z = z_new
    
    time += dt


# reorganising data into dictionary 
data = np.array(data)
data = {
    'time': data[:, 0],
    'x': data[:, 1],
    'y': data[:, 2],
    'z': data[:, 3]
}

# plotting 
fig = plt.figure(figsize = (17,17))
ax = plt.axes(projection='3d')
ax.plot3D(data["x"], data["y"], data["z"], 'b')

ax.set_xlim(x_range)
ax.set_ylim(y_range)
ax.set_zlim(z_range)
ax.set_xticks([-15, -10, -5, 0, 5, 10, 15, 20])
ax.set_zticks([0, 10, 20, 30, 40])
ax.grid(visible=None)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
